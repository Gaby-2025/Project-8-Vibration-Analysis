```python
import serial    # after    pip install pyserial
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq       # after     pip install scipy

fig, (ax1, ax2) = plt.subplots(1, 2)
plt.subplots_adjust(right=2, top=1, wspace=0.25)

arduino_port = "COM10"     # CONFIRM THIS!!
baud = 115200 
ser = serial.Serial(arduino_port, baud)    # resets Arduino
print("Connected to Arduino port:" + arduino_port)

samples = 500 #how many samples to collect
line = 0 #start at 0 because our header is 0 (not real data)

t = []
x = []

while line <= samples:      # collect the samples
    
    getData=ser.readline()
    data = getData.decode('utf-8')  # try/except
    
    if (data.find(",") != -1):
        tdat, xdat = data.split(",")
        t.append(int(tdat))
        x.append(int(xdat.rstrip()))
        line = line + 1
    
ser.close()

ax1.plot(t, x, '.-', color='lightskyblue')
ax1.grid('on')
ax1.set_xlabel('t (ms)', fontsize=14)
ax1.set_ylabel('mma.x', fontsize=14)
plt.title("Motor Speed: 50")

t = np.array(t)
x = np.array(x)
N = t.size
dt = np.mean(np.diff(t))/1000  # time step in seconds
fx = fft(x)
freq = fftfreq(N, dt)[1:N//2]
ax2.plot(freq, 2.0/N * np.abs(fx[1:N//2]))
ax2.grid()
ax2.set_xlabel('f (Hz)', fontsize=14)
ax2.set_ylabel('$|\\hat x|$', fontsize=14);
```
